1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
list=[4,6,9,5,7]
list.append(10)
print("appending:",list)
list.extend([2,3])
print("extending:",list)
list.insert(2,18)
print("inserting:",list)
list.remove(10)
print("removing:",list)
list.pop(2)
print("poping:",list)
list.index(7)
print("indexing:",list)
list.count(3)
print("counting:",list)
list.sort()
print("sorting:",list)
list.reverse()
print("reverse:",list)
list.clear()
print("clearing:",list)
```

5.Write a script that checks if a person is eligible to vote (age ≥ 18). Take age as a variable and print the appropriate message

```python
def vote_eligible():
    age=int(input("Enter your age :"))
    if age>=18:
        print("Your are eligible for voting")
    else:
        print("Your are not eligible for voting")

vote_eligible()
```

```python

def grade(marks):
    if marks>=90:
        return "A"
    elif marks>=75 and marks<=89:
        return "B"
    elif marks>=50 and marks<=74:
        return "C"
    return "Fail"

grade(98)
```

7.Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
def check(num):
    if num>0 and num%2==0:
        print(num,"is positive and even")
    elif num<0 and num%2==0:
        print(num,"is negative and even")
    elif num<0 and num%2!=0:
        print(num,"is negative and odd")
    elif num>0 and num%2!=0:
        print(num,"is positive and odd")
    elif num==0:
        print(num,"is zero")

check(23)
    
    
        
```
