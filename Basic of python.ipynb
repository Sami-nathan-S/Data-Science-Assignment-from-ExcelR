```python
        
def check():
    try:
        num=int(input("Enter a number:"))
        if num <= 1:
           print("Not prime number")

        else:
        
          for i in range(2, num):
               if num % i == 0:
                  print("Not prime number")
                  break
          else:
            print("Prime number")

    
    except ValueError:
            print("Invalid input. Please enter a valid integer.")
    

check()              
              
        
        
```

```python
import random

def product():
    num1=random.randint(1,19)
    num2=random.randint(13,20)
    

    try:
        product=int(input(f"Enter answer for {num1} x {num2}:"))
        if num1*num2==product:
            print(f"Congratulation Correct answer {num1} x {num2} = {product}")
        else:
            print(f"Sorry not Correct answer {num1} x {num2} != {product}")
            print(f"Correct answer is {num1} x {num2} =",num1*num2)
    except ValueError:
            print("Invalid input. Please enter a valid integer.")

product()

            
```
