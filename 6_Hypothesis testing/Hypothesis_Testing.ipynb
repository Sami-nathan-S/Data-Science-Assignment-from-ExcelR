Objective:

To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.

Data Provided: The theoretical weekly operating cost model: W =  1,000+ 5X Sample of 25 restaurants with a mean weekly cost of Rs. 3,050 Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units



HYPOTHESIS STATEMENT:

->Restaurant owners claim costs are higher than the model predicts, so this is a right-tailed test.

-> NULL HYPOTHESIS: mu = 1000+5(600)=4000 mu <= 4000

->Alternative Hypothesis: mu > 4000

Calculate the Test Statistic
● ˉxˉ = sample mean weekly cost (Rs. 3,050) ● μ = theoretical mean weekly cost according to the cost model (W = 1,000+5X for X = 600 units) ● σ = 5*25 units ● n = sample size (25 restaurants)

Step 1: Theoretical Mean Weekly Cost mu=1000+5×600=4000 mu=4000

Step 2: Standard Deviation of Weekly Cost sigma = 5*25 =125 sigma=125

```python
#Step 3: Z-Test Formula
#   Z = X_bar - mu / (sigma/sqrt(n))

    #   Z = 3050−4000 / (125/sqrt(25)

    #  Z = -950 / 25

    # Z = -38
```

```python
#Step 4:  Probability & Critical Value 

#Critical value (right-tailed, alpha = 0.05):


   #  Z(critical) = 1.645
    # Z =-38
```

```python
#Step 5 : Compare Z value and the critical value

    #Reject H0:
                # z > 1.645
    #Accept H0:
               #  −38 < 1.645
```

```python
 #  Therfore we Fail to reject the null hypothesis
```

```python
import math
from scipy.stats import norm

# Given values
sample_mean = 3050
X_mean = 600
sigma_X = 25
n = 25
alpha = 0.05

# population mean cost
mu = 1000 + 5 * X_mean

# Standard deviation of weekly cost
sigma = 5 * sigma_X

# Z-test calculation
z = (sample_mean - mu) / (sigma / math.sqrt(n))

# Critical value (right-tailed test)
z_critical = norm.ppf(1 - alpha)

print("Hypothesis Testing Results")
print("----------------------------")
print("Theoretical Mean Cost (μ):", mu)
print("Sample Mean Cost (x̄) : " ,sample_mean)
print("Z-statistic : " ,z)
print("Critical Z-value:",z_critical)

if z > z_critical:
    print(" Reject the Null Hypothesis")
    print(" Weekly operating costs are higher than the model predicts.")
else:
    print(" Fail to Reject the Null Hypothesis")
    print(" No evidence that weekly operating costs are higher than the model.")
```

```python
 # CONCLUSION :

#There is no statistical evidence to support the restaurant owners.
# This claims that weekly operating costs are higher than the theoretical model
#  Therfore we Fail to reject the null hypothesis
```
