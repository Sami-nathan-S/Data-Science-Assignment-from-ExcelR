Problem Statement : Hospital Patient Data Analysis

Context:

A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. You have two datasets: one with patient info and another with billing details. Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.

Tasks:

1.	Load the patient dataset and show summary with info().
  
3.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
  
5.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

  
7.	Use groupby to find total bill amount per department.

  
9.	Remove duplicate patient records based on PatientID.

   
11.	Fill missing BillAmount values with the mean bill amount.

   
13.	Merge the billing dataset with patient dataset on PatientID.

   
15.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

   
17.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

   
Expected Outcome:


•	Final cleaned dataset with accurate billing info.

•	All missing values handled, merged dataset across PatientID.

•	Ability to perform further analytics on department-wise revenue or doctor performance.


```python
import pandas as pd
```

```python
patient_df=pd.read_csv(r"C:\Users\Sami\Downloads\Patient_Data.csv")
billing_df=pd.read_csv(r"C:\Users\Sami\Downloads\Billing_Data.csv")
```

```python
patient_df.info()
```

```python
bill=patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
bill
```

```python
patient_df = patient_df.drop(['ReceptionistID', 'CheckInTime'], axis=1)

print(patient_df.head())
```

```python
dep_bill=patient_df.groupby('Department')['BillAmount'].sum()
dep_bill
```

```python
patient_df=patient_df.drop_duplicates(subset='PatientID')
patient_df
```

```python
mean_bill=patient_df['BillAmount'].mean()
patient_df['BillAmount'].fillna(mean_bill)
```

```python
merged_df = pd.merge(patient_df, billing_df, on='PatientID')
merged_df
```

```python
new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Name':['Kavee','Sriman'],
    'Department': ['Cardiology', 'Orthopedics'],
    'Doctor': ['Dr. John', 'Dr. Lee'],
    'BillAmount': [4500, 3800]
})
new_patients
```

```python
new_patients_df = pd.DataFrame(new_patients)
merged_df = pd.concat([patient_df,new_patients_df],axis=0,ignore_index=True)
merged_df
```

```python
merged_final_df = pd.concat([merged_df, billing_df[['InsuranceCovered', 'FinalAmount']]], axis=1)
merged_final_df['InsuranceCovered'] = merged_final_df['InsuranceCovered'].fillna(merged_final_df['InsuranceCovered'].mean())
merged_final_df['FinalAmount'] = merged_final_df['FinalAmount'].fillna(merged_final_df['FinalAmount'].mean())
merged_final_df
```

Outcome

Final cleaned dataset with accurate billing info.
